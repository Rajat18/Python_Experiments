### Normal Distribution

Wrintg code to get a random normal distribution

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
# Getting random array
x_min = 0.0
x_max = 16.0
x = np.linspace(x_min, x_max, 100)
x1 = np.linspace(12, 22, 100)
print(x)
print(x1)
```

```python
#Using scipy.stats to generate random distribution
import scipy.stats

mean = 8.0 
std = 2.0

y = scipy.stats.norm.pdf(x,mean,std)
print(y)
y1 = scipy.stats.norm.pdf(x1,16,4)
print(y1)
```

```python
x_prod = x * x1
print(x_prod)
y_prod = y * y1
print(y_prod)
```

```python
#Plotting the normal distributino
import matplotlib.pyplot as plt

plt.plot(x_prod,y_prod, color='coral')
plt.grid()
# plt.xlim(x_min,x_max)
# plt.ylim(0,0.25)
plt.title('How to plot a normal distribution in python with matplotlib',fontsize=10)
plt.xlabel('x')
plt.ylabel('Normal Distribution')
# plt.savefig("normal_distribution.png")
plt.show()
```

```python
# Another way to generate normal distribution
mu, sigma = 0.5, 0.1
s = np.random.normal(mu, sigma, 1000)

# Create the bins and histogram
count, bins, ignored = plt.hist(s, 20, normed=True)

# Plot the distribution curve
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
    np.exp( - (bins - mu)**2 / (2 * sigma**2) ),       linewidth=3, color='y')
plt.show()
```

### Calculating the Z Score and p value

Using sciPy stats.zscore() function | Python

```python
#Calculating the Z Scores
# stats.zscore() method   
import numpy as np 
from scipy import stats 
    
arr1 = [[20, 2, 7, 1, 34], 
        [50, 12, 12, 34, 4]] 
  
arr2 = [[50, 12, 12, 34, 4],  
        [12, 11, 10, 34, 21]] 
  
print ("\narr1 : ", arr1) 
print ("\narr2 : ", arr2) 
```

```python
z_scores = stats.zscore(arr1, axis = 1)
print ("\nZ-score for arr1 : \n", z_scores) 
```

```python
p_values = scipy.stats.norm.sf(abs(z_scores)) #one-sided
print(p_values)
p_values = scipy.stats.norm.sf(abs(z_scores))*2 #twosided
print(p_values)
```

```python
z_scores = stats.zscore(arr2, axis = 1)
print ("\nZ-score for arr1 : \n", z_scores) 
```

```python
p_values = scipy.stats.norm.sf(abs(z_scores)) #one-sided
print(p_values)
p_values = scipy.stats.norm.sf(abs(z_scores))*2 #twosided
print(p_values)
```
